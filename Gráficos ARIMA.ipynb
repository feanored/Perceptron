```python
import numpy as np
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
import warnings
warnings.simplefilter("ignore", UserWarning)
warnings.simplefilter("ignore", FutureWarning)
warnings.filterwarnings("ignore")
```

```python
y_10 = sm.tsa.arma_generate_sample(ar=np.r_[1,-1], ma=np.r_[1, 0], nsample=500)
model = sm.tsa.ARMA(y_10, (1,0)).fit(trend='nc', disp=0)
model.params
```

```python
y_01 = sm.tsa.arma_generate_sample(ar=np.r_[1,0], ma=np.r_[1,1], nsample=500)
model = sm.tsa.ARMA(y_01, (0, 1)).fit(trend='nc', disp=0)
model.params
```

```python
y_11 = sm.tsa.arma_generate_sample(ar=np.r_[1,-1], ma=np.r_[1,1], nsample=500)
model = sm.tsa.ARMA(y_11, (1, 1)).fit(trend='nc', disp=0)
model.params
```

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 14))
lags = 25

acf_10 = plot_acf(y_10, lags=lags, ax=ax[0][0], title="FAC AR(1)")
pacf_10 = plot_pacf(y_10, lags=lags, ax=ax[0][1], title="FACP AR(1)")

acf_01 = plot_acf(y_01, lags=lags, ax=ax[1][0], title="FAC MA(1)")
pacf_01 = plot_pacf(y_01, lags=lags, ax=ax[1][1], title="FACP MA(1)")

acf_11 = plot_acf(y_11, lags=lags, ax=ax[2][0], title="FAC ARMA(1, 1)")
pacf_11 = plot_pacf(y_11, lags=lags, ax=ax[2][1], title="FACP ARMA(1, 1)")

plt.show()
```
